### κ²°μΈ΅μΉ μ κ±°

```python
# λ³΄μ¦κΈ, μ›”μ„Έ, κ¶λ¦¬κΈμ΄ 0μΈ λ°μ΄ν„° ν™•μΈ
df_μ„μΈν‰μμ „μ›”μ„Έ.loc[(df_μ„μΈν‰μμ „μ›”μ„Έ['λ³΄μ¦κΈ']==0) | (df_μ„μΈν‰μμ „μ›”μ„Έ['κ¶λ¦¬κΈ']==0) | (df_μ„μΈν‰μμ „μ›”μ„Έ['μ›”μ„Έ']==0)]
```

```python
# λ³΄μ¦κΈ, μ›”μ„Έ, κ¶λ¦¬κΈμ΄ 0μΈ λ°μ΄ν„° μ κ±°, μΈλ±μ¤ μ΄κΈ°ν™”
idx = df_μ„μΈν‰μμ „μ›”μ„Έ.loc[(df_μ„μΈν‰μμ „μ›”μ„Έ['λ³΄μ¦κΈ']==0) | (df_μ„μΈν‰μμ „μ›”μ„Έ['κ¶λ¦¬κΈ']==0) | (df_μ„μΈν‰μμ „μ›”μ„Έ['μ›”μ„Έ']==0)].index
df_μ„μΈν‰μμ „μ›”μ„Έ.drop(idx, inplace=True)
df_μ„μΈν‰μμ „μ›”μ„Έ.reset_index(drop=True, inplace=True)
```
